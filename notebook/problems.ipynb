## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np
import numpy.random as rnd
import scipy as scp
import sympy as syp
import matplotlib.pyplot as plt

## Run the experiment

rng = rnd.seed(720) # For repeatibility

dieRoll    = rnd.randint(1,7,(1000,2)) # Define roll of two dice a 1000 times
sampleSize = len(dieRoll) # Define sample set

c = 0 # Initialize counter
for pos in range(0,sampleSize):
    if dieRoll[pos,0]+dieRoll[pos,1]==7:
        c += 1

empiricalProb = c/sampleSize
print(f'The probability of two dice rolling to add up to exactly 7 is {empiricalProb*100:0.2f}%')

## Exact probability

# Two dice will add up to 7 for the following pairs (6,1), (5,2), (4,3), (3,4), (2,5), (1,6)
# in a sample set of 6x6 = 36 total possibilities. Thus:

exactProb = 1/6
print(f'The probability of two dice rolling to add up to exactly 7 is {exactProb*100:0.2f}%')

```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
randPick = rnd.randint(1,4,(1,5))
randPick
print(randPick[0][0])
```

```python
# Define the ball box

ballWhite   = np.full(10,'white')
ballRed     = np.full(20,'red')
ballGreen   = np.full(30,'green')

ball_box    = np.concatenate((ballWhite,ballRed),axis=0)
ball_box    = np.concatenate((ball_box,ballGreen),axis=0)

# Define the sample size

sampleSize  = 1000
            
# Start random pick
c1          = 0 # Start counter
c2          = 0 # Start counter

# Start the experiments
for pick in range(0,sampleSize):
    choiceColor = []
    for randPos in range(5):
        randPick = rnd.choice(ball_box)
        choiceColor.append(randPick)
    
    # Count the number of occurrences
    whiteCount  = choiceColor.count('white')
    redCount    = choiceColor.count('red')
    greenCount  = choiceColor.count('green')

    # Get true events
    if (whiteCount == 3) & (redCount == 2):
        c1 += 1
    elif (whiteCount == 5) | (redCount == 5) | (greenCount == 5):
        c2 += 1 

# Compute empirical probabilities
empiricalProb1 = c1/sampleSize
empiricalProb2 = c2/sampleSize

print(f'The probability of picking 3 white balls and 2 red balls is {empiricalProb1*100:0.4f}%')
print(f'The probability of picking all balls of the same color {empiricalProb2*100:0.4f}%')
```
